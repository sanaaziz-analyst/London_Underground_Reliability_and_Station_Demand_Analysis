```python
#automate the same 8 steps for all 6 years at once
years = [2011, 2012, 2013, 2014, 2015, 2016, 2017]
all_years = []

for year in years:
    sheet = f"{year} Entry & Exit"
    df_year = pd.read_excel("multi-year-station-entry-and-exit-figures.xls", sheet_name=sheet, skiprows=6)

    if df_year.shape[1] == 11:
        df_year.columns = ["nlc", "station", "borough", "note",
                            "entry_weekday", "entry_saturday", "entry_sunday",
                            "exit_weekday", "exit_saturday", "exit_sunday",
                            "annual_entry_exit_million"]
    else:
        df_year.columns = ["nlc", "station", "note",
                            "entry_weekday", "entry_saturday", "entry_sunday",
                            "exit_weekday", "exit_saturday", "exit_sunday",
                            "annual_entry_exit_million"]
        df_year["borough"] = None

    df_year = df_year[df_year["nlc"].notna()]
    df_year["year"] = year
    all_years.append(df_year)

df_footfall = pd.concat(all_years, ignore_index=True)
df_footfall.shape
```

```python
#Sanity-check the combined table
df_footfall["year"].value_counts().sort_index()
```

```python
df_footfall[df_footfall["year"] == 2011]["borough"].unique()
```

```python
#Convert the numeric columns properly
count_cols = ["entry_weekday", "entry_saturday", "entry_sunday",
              "exit_weekday", "exit_saturday", "exit_sunday",
              "annual_entry_exit_million"]

for col in count_cols:
    df_footfall[col] = pd.to_numeric(df_footfall[col], errors="coerce")

df_footfall.info()
```

```python
#Find those 5 problem rows
df_footfall[df_footfall["station"].isna()]
```

```python
#Add a second safety filter
df_footfall = df_footfall[df_footfall["station"].notna()]
df_footfall.shape
#it should now be 1881 - 5 = 1876 rows.
```

```python
df_footfall.info()
```

```python
#helpful derived columns
df_footfall["total_weekday"] = df_footfall["entry_weekday"] + df_footfall["exit_weekday"]
df_footfall["total_weekend"] = (df_footfall["entry_saturday"] + df_footfall["exit_saturday"] +
                                  df_footfall["entry_sunday"] + df_footfall["exit_sunday"])
df_footfall.head()
```

```python
#Export the clean file
df_footfall.to_csv("clean_station_footfall.csv", index=False)
```
